```python
df['class']=df['class'].apply(lambda x:1 if x=="Business" else 0)
```

```python
def stops_preprocessing(x):
    if(x=="zero"):
        return 0
    elif(x=="one"):
        return 1
    else:
        return 2
    
```

```python
df['stops']=df['stops'].apply(stops_preprocessing)
```

```python
columns=["airline","source_city","departure_time","arrival_time","destination_city"]
for col in columns:
    value=df[col].value_counts()
    plt.pie(value,labels=value.index)
    plt.show()
```
